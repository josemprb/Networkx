```python
import networkx as nx
import time
import collections
import matplotlib.pyplot as plt
```

```python
n = int(input("Enter a number of nodes for the network: "))
```

```python
net = nx.Graph()
net.clear()

for i in range(1, n):
    net.add_node(0)
    net.add_node(i)
    net.add_edge(0,i)

nx.draw(net)
```

```python
network = nx.read_edgelist('../course/word_association_graph_DSF.txt',
                          create_using = nx.DiGraph(), nodetype = str,
                          data = [('weight', float), ])
```

I did an implementation that would last about 1 hour and 10 minutes of execution time.

```python
dictionary = {}
start = time.time()

for i in range(1000):
    dictionary[list(network.edges())[i][0]] = dictionary.get(list(network.edges())[i][0], 0) + 1 
    dictionary[list(network.edges())[i][1]] = dictionary.get(list(network.edges())[i][1], 0) + 1 

print(dictionary)
end = time.time()
print(str(end - start))
```

```python
degree_map = collections.defaultdict(int)
start = time.time()

for nodeA, nodeB in network.edges():
    degree_map[nodeA] += 1
    degree_map[nodeB] += 1

print(degree_map)
end = time.time()
print(str(end - start))
```

```python
in_degree_map = collections.defaultdict(int)
out_degree_map = collections.defaultdict(int)
start = time.time()

for nodeA, nodeB in network.edges():
    in_degree_map[nodeB] += 1
    out_degree_map[nodeA] += 1

print(in_degree_map)
print("I need some space, please.")
print(out_degree_map)
end = time.time()
print(str(end - start))
```

```python
indegree_map = {node:0 for node in network.nodes()}
outdegree_map = {node:0 for node in network.nodes()}

for source, target in network.edges():
    outdegree_map[source] += 1
    indegree_map[target] += 1

fig, axs = plt.subplots(2)
axs[0].hist(indegree_map.values())
axs[1].hist(outdegree_map.values())
```
